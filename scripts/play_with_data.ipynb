```python
import pandas as pd
from asodesigner.consts import EXPERIMENT_RESULTS

all_features = pd.read_csv(EXPERIMENT_RESULTS / 'Angptl2' / 'all_features.csv')
# all_features
```

```python
cols['antisense'] = cols[SEQUENCE]
merged_df = pd.merge(cols, all_features, on='antisense', how='inner')
print(list(cols['antisense'])[0])
print(list(all_features['antisense'])[0])
merged_df.head(2)
```

```python
from scipy.stats import pearsonr

features = ['gc_content', 'on_target_energy_max', 'on_target_energy_sum', 'on_target_energy_fits', 'on_target_fold_openness', 'on_target_fold_openness_normalized', 'mfe', 'aso_aso_tm', 'aso_aso_delta_g', 'aso_aso_delta_h', 'aso_aso_delta_s', 'total_hybridization_max_sum_human', '3_matches_human', '2_matches_human', '1_matches_human', '0_matches_human', 'total_hybridization_max_sum_yeast', '0_matches_yeast', '1_matches_yeast', '2_matches_yeast', '3_matches_yeast']
for feature in features:
    corr, p_value = pearsonr(merged_df[INHIBITION], merged_df[feature])
    print(f"Feature: {feature:<35}, Pearson: {corr:<5.2f}, p-value: {p_value:<10.2} ")
```

```python
from scipy.stats import spearmanr

for feature in features:
    corr, p_value = spearmanr(merged_df[INHIBITION], merged_df[feature])
    print(f"Feature: {feature:<35}, Spearman: {corr:<5.2f}, p-value: {p_value:<10.2} ")
```
