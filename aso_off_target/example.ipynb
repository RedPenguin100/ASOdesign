```python
pre_mrna_df = pd.read_csv('pre_mRNA.csv')
pre_mrna_dic = pre_mrna_df.set_index("gene")["seq"].to_dict()
```

```python
exmaple1 = pre_mrna_dic['KRAS'][41212:41228]
```

```python
def reverse_complement(seq: str) -> str:
    complement = str.maketrans("ACGTNacgtn", "TGCANtgcan")
    return seq.translate(complement)[::-1]

# Example
assert(reverse_complement(exmaple1)=='GCTAAAACAAATGCTA')  
```
